```python
# Исходная функция в двумерном пространстве
def f(x, y):
    return x**2+y**2

def gradient(x, y):
    step = 0.00001 # шаг дифференцирования
    df_dx = (f(x+step, y) - f(x-step, y)) / (2*step) # Центральная разность 
    df_dy = (f(x, y+step) - f(x, y-step)) / (2*step) # Центральная разность
    return [df_dx, df_dy] # Возвращаем вектор

x, y = 2, 3 
res = gradient(x, y)
print(f"Градиент в точке ({x}, {y}):", res)
```

```python
# Исходная функция в трёхмерном пространстве
def f(x, y, z):
    return x**2+y**2+z**2

def gradient(x, y, z):
    step = 0.00001 # шаг дифференцирования
    df_dx = (f(x+step, y, z) - f(x-step, y, z)) / (2*step) # Центральная разность 
    df_dy = (f(x, y+step, z) - f(x, y-step, z)) / (2*step) # Центральная разность
    df_dz = (f(x, y, z+step) - f(x, y, z-step)) / (2*step) # Центральная разность
    return [df_dx, df_dy, df_dz] # Возвращаем вектор

x, y, z = 2, 3, 4
res = gradient(x, y, z)
print(f"Градиент в точке ({x}, {y}, {z}):", res)
```

```python
# Исходная функция в двумерном пространстве
def F(x, y):
    return (-x**5, 3*y)

def rotor(x, y):
    step = 0.00001

    Fy_dx_R = F(x+step, y)[1]
    Fy_dx_L = F(x-step, y)[1]
    Fy_dx = (Fy_dx_R - Fy_dx_L) / (2*step)

    Fx_dy_R = F(x, y+step)[0]
    Fx_dy_L = F(x, y-step)[0]
    Fx_dy = (Fx_dy_R - Fx_dy_L) / (2*step)

    result = Fy_dx - Fx_dy

    return result # Возвращаем скаляр

x, y = 2, 3
res = rotor(x, y)
print(f"Ротор в точке ({x}, {y}): {res}")
```

```python
# Исходная функция в трёхмерном пространстве
def F(x, y, z):
    return (-y, x, 0)

def rotor(x, y, z):
    step = 0.00001 # шаг дифференцирования

    Fx_dz_R = F(x, y, z + step)[0]
    Fx_dz_L = F(x, y, z - step)[0]
    Fx_dz = (Fx_dz_R - Fx_dz_L) / (2 * step)

    Fy_dz_R = F(x, y, z + step)[1]
    Fy_dz_L = F(x, y, z - step)[1]
    Fy_dz = (Fy_dz_R - Fy_dz_L) / (2 * step)

    Fz_dy_R = F(x, y + step, z)[2]
    Fz_dy_L = F(x, y - step, z)[2]
    Fz_dy = (Fz_dy_R - Fz_dy_L) / (2 * step)

    Fz_dx_R = F(x+step, y, z)[2]
    Fz_dx_L = F(x-step, y, z)[2]
    Fz_dx = (Fz_dx_R - Fz_dx_L) / (2 * step)

    Fy_dx_R = F(x+step, y, z)[1]
    Fy_dx_L = F(x-step, y, z)[1]
    Fy_dx = (Fy_dx_R - Fy_dx_L) / (2 * step)   

    Fx_dy_R = F(x, y+step, z)[0]
    Fx_dy_L = F(x, y-step, z)[0]
    Fx_dy = (Fx_dy_R - Fx_dy_L) / (2 * step)

    rot_x = Fz_dy - Fy_dz
    rot_y = Fz_dx - Fx_dz
    rot_z = Fy_dx - Fx_dy

    return (rot_x, rot_y, rot_z)

x, y, z = 2, 3, 1
res_rotor = rotor(x, y, z)
print(f"Ротор в точке ({x}, {y}, {z}):", res_rotor)
```

```python
# Исходная функция в двумерном пространстве
def F(x, y):
    return (-3*(y**2), 3*x)

def divergent(x, y):
    step = 0.00001

    Fy_dx_R = F(x+step, y)[1]
    Fy_dx_L = F(x-step, y)[1]
    Fy_dx = (Fy_dx_R - Fy_dx_L) / (2*step)

    Fx_dy_R = F(x, y+step)[0]
    Fx_dy_L = F(x, y-step)[0]
    Fx_dy = (Fx_dy_R - Fx_dy_L) / (2*step)

    result = Fy_dx + Fx_dy

    return result # Возвращаем скаляр

x, y = 2, 3
res = divergent(x, y)
print(f"Дивергент в точке ({x}, {y}): {res}")
```

```python
# Исходная функция в трехмерном пространстве
def F(x, y, z):
    return (-3 * (y**2), 3 * x, z)
def divergent(x, y, z):
    step = 0.00001  # Шаг дифференцирования

    Fx_dx_R = F(x+step, y, z)[0] 
    Fx_dx_L = F(x-step, y, z)[0]
    Fx_dx = (Fx_dx_R - Fx_dx_L) / (2 * step)

    Fy_dy_R = F(x, y+step, z)[1]
    Fy_dy_L = F(x, y-step, z)[1]
    Fy_dy = (Fy_dy_R - Fy_dy_L) / (2 * step)

    Fz_dz_R = F(x, y, z+step)[2]
    Fz_dz_L = F(x, y, z-step)[2]
    Fz_dz = (Fz_dz_R - Fz_dz_L) / (2 * step)

    result = Fx_dx + Fy_dy + Fz_dz

    return result  # Возвращаем скаляр

x, y, z = 2, 3, 1
res = divergent(x, y, z)
print(f"Дивергенция в точке ({x}, {y}, {z}): {res}")
```
